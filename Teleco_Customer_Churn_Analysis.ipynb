```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Customer Churn.csv')
df.head(10)
```

```python
df.info()
```

```python
# Replacing missing values to "0", and changing datatype to float
```

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ", "0")
df["TotalCharges"] = df["TotalCharges"].astype(float)
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

```python
def conv(value):
  if value == 1:
    return "yes"
  else:
    return "no"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
df.head(10)
```

```python
# Converted 0 and 1 values to yes/no for better understanding
```

```python
df.columns
```

```python
ax = sns.countplot(x ="Churn", data = df)
ax.bar_label(ax.containers [0])
plt.title("Churn Distribution by Count")
plt.show()
```

```python
gb = df.groupby("Churn").agg({'Churn':'count'})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.title("Churn Distribution by Percentage")
plt.show()
```

```python
### From the above pie chart we can see that 26.54% of the customers churned out
### Now lets understand the reason behind it.
```

```python
sns.countplot(x = "gender", data = df, hue = "Churn")
plt.title("Gender Distribution by Churn")
plt.show()
```

```python
# The above graph shows that the Churn is not affect by the gender.
```

```python
sns.countplot(x = "SeniorCitizen", data = df, hue = "Churn")
plt.title("Gender Distribution by Senior Citizen")
plt.show()
```

```python
ct = pd.crosstab(df["SeniorCitizen"], df["Churn"], normalize='index') * 100

# 2. Plot the stacked bar chart
ax = ct.plot(kind='bar', stacked=True)

# 3. Add title and labels
plt.title("Senior Citizen vs Churn (%)")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage")

# 4. Add percentage labels on the bars
for idx, row in ct.iterrows():
    cumulative = 0
    for churn_value in ct.columns:
        value = row[churn_value]
        if value > 0:
            # Use the numerical location of the index for x-position
            x_pos = ct.index.get_loc(idx)
            plt.text(
                x_pos,                       # x-position
                cumulative + (value / 2),  # y-position
                f'{value:.1f}%',           # label text
                ha='center', va='center'
            )
        cumulative += value

plt.legend(title="Churn")
plt.show()
```

```python
# Comparitavely a greater percentage of people from Senior citizen have churned.
```

```python
plt.figure(figsize = (10, 5))
sns.histplot(x = "tenure", data = df, bins = 72, hue = "Churn")
plt.title("Tenure Distribution by Churn")
plt.show()
```

```python
# People who have used the services for longer period of time have stayed. And people who have used services for a month or two have churned.
```

```python
ax = sns.countplot(x = "Contract", data = df, hue = "Churn")
ax.bar_label(ax.containers[0])
plt.title("Count Of Customers by Contract")
plt.show()
```

```python
# People who have month to month contract are likely to churn than those who have 1 or 2 years of contract.
```

```python
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

plt.figure(figsize=(15, 15))

for i, col in enumerate(cols, 1):
    plt.subplot(3, 3, i)   # 3 rows, 3 columns
    sns.countplot(data=df, x=col, hue="Churn")
    plt.title(f"Countplot of {col} by Churn")
    plt.xticks(rotation=45)
    plt.xlabel('')
    plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# A large portion of churn comes from customers who have internet services (DSL or Fiber), especially those who do not subscribe to value-added services.
# Streaming services (TV/Movies) also show higher churn among customers who use them but lack security or support add-ons.
```
